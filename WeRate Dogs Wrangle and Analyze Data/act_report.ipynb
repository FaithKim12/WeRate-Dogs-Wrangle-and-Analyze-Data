### I wanted to go further and get insights on sources that generate the highest and lowets tweets. So I plotted a horizontal bar plot with the help of Matplotlib. From the plot above, we can see that over 1900 tweets which is 90%+ were posted from Iphone. The lowest number of Tweets came from TweetDeck. We can therefore conclude that dog lovers are more likely to use an Iphone.
